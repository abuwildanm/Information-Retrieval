## Skenario 1 - Tanpa Stemming Dengan Stopword

```
# Import Library Standard
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
# Import Library Sklearn
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.naive_bayes import MultinomialNB
```

```
# Read Dataset
data = pd.read_csv('data_sms_spam.csv')
data.head()
```

```
# Ukuran Dataset
print('Ukuran dataset:', data.shape)
```

```
# Random urutan dan split data training & testing
x_train, x_test, y_train, y_test = train_test_split(data['teks'], data['label'], test_size=0.4)
```

```
train_data = pd.DataFrame({'teks':x_train, 'label':y_train})
test_data = pd.DataFrame({'teks':x_test, 'label':y_test})
```

```
train_data.head()
```

```
test_data.head()
```

```
# Ukuran Data Training & Testing
print('Ukuran data train:', train_data.shape)
print('Ukuran data test:', test_data.shape)
```

```
# Perhitungan bobot kata
vectorizer = TfidfVectorizer()
bobot = vectorizer.fit_transform(train_data['teks'])
print('Ukuran matriks bobot:', bobot.shape)
```

```
print('Daftar Term:')
vectorizer.get_feature_names()
```

```
print('Daftar Stopword:')
vectorizer.get_stop_words()
```

```
# Perhitungan cosine similarity
similarity_matrix = cosine_similarity(bobot, bobot)
print('Ukuran matriks cosine similarity: ', similarity_matrix.shape)
```

```
print('Matriks Cosine Similarity:')
similarity_matrix
```
